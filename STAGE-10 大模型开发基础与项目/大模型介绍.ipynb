# 什么是语言模型？
- 定义：判断一句话发生的概率（你可以把语言模型想象成一个“猜词游戏”高手）
- 标准：假如一个句子为s=[w1, w2, w3, ··· wn]语言模型是计算当前句子发生的概率，即：P(s)=P(w1, w2, w3, ··· wn)如果符合人类日常用语表达，P(s)就更大，否则就是小
- 分类：

N-gram 是“统计词组”，神经网络模型是“理解词义”，Transformer 是“理解语言与推理”。
