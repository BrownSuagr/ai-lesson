# 1、什么是LangChain？
- LangChain由Harrison Chase创建于2022年10月它是围绕LLMs（大语言模型）建立的一个框架，LangChain自身并不开发LLMs，它的核心理念是为各种LLMs实现通用的接口，把LLMs相关的组件“链接”在一起，简化LLMs应用的开发难度，方便开发者快速地开发复杂的LLMs应用。


# 2、LangChain现在由几种语言的实现？
- Python
- Nodejs

# 3、LangChain有哪些主要组件？
- Models：模型，各种类型的模型和模型集成，如GPT-4
- Prompts：提示，包括提示管理、提示优化和提示序列化
- Memory：记忆，用来保存和模型交互时的上下文状态
- Indexs：索引，用结构化文档，以便于模型的交互
- Chains：链，一些列对各种组件调用
- Agents：代理，决定模型采取哪些行动，执行并观察流程，直到完成为止

# 4、目前LangChain支持Model哪几种类型，每种类型有哪些特点？
- LLMs：大语言模型接受文本字符作为输入，返回也是文本字符
- 聊天模型：基于LLMs。不同得失它接受聊天消息（一种特定格式数据），返回的也是聊天消息
- 文本嵌入模型：文本嵌入模型接受文本作为输入，返回的是浮点数列表


# 5、LangChain有哪些应用场景？
- 个人助手
- 基于文档的问答系统
- 聊天机器人
- Tabular数据查询
- API交互
- 信息提取
- 文档总结
