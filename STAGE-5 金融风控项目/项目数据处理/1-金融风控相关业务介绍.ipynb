# 金融风控相关业务学习目标
- 知道常见的信贷风险
- 知道机器学习风控模型的优势
- 知道信贷领域常用的属于含义


# 1、什么事信贷业务？
信贷业务就是贷款业务，是商业银行和互联网金融公司最重要的资产业务和主要的盈利手段
- 通过放款收回本金和利息，扣除成本后获取利润
- 贷款平台预测有信贷需求的用户的还款情况，然后将本金借贷给还款概率大的用户

# 2、常见的信贷业务中有哪些风险？
- 小额贷款无抵押物，存在不能及时还款风险
- 对于预测信用较差的人，通过提高利润或者避免放贷减少风险

# 3、信贷领域的两类风险是什么？
- 信用风险：借款人的还款能力和还款意愿出现问题的风险
- 欺诈风险：借款人没想还钱，以欺诈为目的

# 4、风控业务主要针对是那两类风险？
- 信用评分系统：针对信用风险
- 反欺诈系统：针对欺诈风险

# 5、基于机器学习的人工智能风控模型对比传统的人工审批优缺点？
- 人工审批：效率低，对业务能力要求较高，不适合金融零售业务场景
- 机器学习模型：批量，迅速，精准、同时可以处理大量贷款请求

# 6、风控业务中有那些专业术语？
- DPD：day past due逾期天数 DPD0到期当日，DPD1为逾期一天，DPD7为逾期一周
- FPD：first time past due首次逾期天数
- F/S/T/QPD：首次/二次/三次，四次逾期天数
- M1：逾期[1, 30]天M是英文month首字母意思
- M1+：逾期[30, inf]天
- default：坏账（动词）
- delinquency：拖欠
- flow rate：流动率，一般指M1向M2，M2向M3转移的比例
- bad rate：坏账率，当月不良资产数/总资产数
- vintage：账龄分析

# 7、常见信贷产品有哪些分类？
- 单期产品
- 多期产品
- 循环额度产品

# 8、常见信贷产品还款方式有哪些？
- 砍头息（短期产品）
- 等额本金
- 等额本息

# 9、常见信贷产品有哪些风险？
- 冒名顶替，黑产骗贷
- 多头借贷，借新还旧
- POS机套现
- 针对风控模型，制作数据


# 小结：
- 知道常见的信贷风险
    - 信用风险
    - 欺诈风险
- 知道机器学习风控模型的优势
    - 批量、迅速、准确、同事处理大量贷款请求
    - 在零售信贷业务业务场景下，与人工审核相比优势显著
- 知道信贷领域常用术语含义
    - DPD逾期天数
    - FPD首次预期天数
    - bad rate 坏账率
    - M1逾期一个月以内

```python
import pandas as pd
import datetime
from pyecharts.charts import *
from pyecharts import options as opts

df1 = pd.read_excel('./file/业务数据.xls')
df2 = df1.copy()
df2.head()
```

```python
'''
    数据条目：5257
    字段数目：14
    数据类型：float64(4), int64(1), object(9)
    存在空的字段：开票金额、实收金额、未收金额、预计付款日、实际到账日、开票日期
    使用内存：575.1+ KB
'''
df2.info()
```

```python
df2.describe()
```

```python
# 数据处理，填充缺失值，将日期类型转换为datetime类型
today_time = pd.to_datetime(df2['实际到账日'].fillna('0').max())

df2['实收金额'] = df2['实收金额'].fillna(0)
df2['开票金额'] = df2['开票金额'].fillna(0)
df2['未收金额'] = df2['未收金额'].fillna(0)

df2['账单周期'] = pd.to_datetime(df2['账单周期'])
df2['应付日期'] = pd.to_datetime(df2['应付日期'])

df2['实际到账日'] = pd.to_datetime(df2['实际到账日']).fillna(today_time)
```

```python
# 在数据原始基础上构造新的字段
df2['是否到期'] = df2.apply(lambda x : 0 if x['应付日期'] > today_time else 1, axis = 1)
df2['是否到期90天'] = (today_time -df2['应付日期']).map(lambda x : 1 if x.days >= 90 else 0)
df2['未收金额2'] = (df2['账单金额'] - df2['实收金额'])
df2['历史逾期天数'] = df2.apply(lambda x : (x['实际到账日'] - x['应付日期']).days if x['未收金额2'] == 0 else (today_time - x['应付日期']).days, axis = 1)
df2['当前逾期天数'] = df2.apply(lambda x : (x['历史逾期天数']) if x['未收金额2'] > 0 else 0, axis = 1)
```

```python
df3 = df2.copy()
df3['账单季度'] = df3['账单周期'].map(lambda x : x.to_period("Q"))
condition = (df3['账单季度'] <= '2018Q4') & (df3['账单季度'] >= '2017Q3')
df3 = df3[condition]
df3.shape
```

```python
# 账单金额
fn1 = df3.groupby('账单季度')[['账单金额']].sum()
fn1.columns = ['账单金额']
fn1
```

```python
# 90到期金额
condition = df3['是否到期90天'] == 1
df4 = df3[condition]
fn2 = df4.groupby('账单季度')[['账单金额']].sum()
fn2.columns = ['到期金额']
fn2
```

```python
df4 = df3[condition]
fn3 = df4.groupby('账单季度')[['未收金额2']].sum()
fn3.columns = ['当前逾期90+金额']
fn3
```

```python
# 合并数据计算逾期率
dfs = [fn1, fn2, fn3]
final1 = pd.concat(dfs, axis = 1)
final1
```

```python
final1['90+净坏账率'] = round(final1['当前逾期90+金额'] / final1['到期金额'], 3)
final1
```

```python
# pyecharts绘图
x_index_list = list(final1.index.values.astype(str))
opt_title = opts.TitleOpts(title = "90+净坏账率")
y_axis = opts.AxisOpts(
    name = "90+净坏账率",
    type_ = "value",
    min_ = 0,
    max_ = 0.014,
    position = "right",
    axisline_opts = opts.AxisLineOpts(linestyle_opts = opts.LineStyleOpts(color="#d14a61")),
    axislabel_opts = opts.LabelOpts(formatter = "{value}"),
)
bar = (
    Bar().add_xaxis(x_index_list)
         .add_yaxis(
            '账单金额', 
            list(final1['账单金额']), 
            yaxis_index = 0, 
            color = '#5793f3'
          )
         .set_global_opts(title_opts = opt_title)
         .extend_axis(yaxis = y_axis)
)
line = (
    Line()
    .add_xaxis(list(final1.index.values.astype(str)))
    .add_yaxis(
        "90+净坏账率",
        list(final1['90+净坏账率']),
        yaxis_index = 1, 
        color = "#675bba",
        label_opts = opts.LabelOpts(is_show=False),
    )
)
bar.overlap(line).render("./file/90+净坏账率.html")
# bar.overlap(line).render_notebook()
```

```python
# 60天账期的账单金额
df4 = df3[(df3['账期'] == 60) & (df3['是否到期'] == 1)]
fn1 = df4.groupby('账单季度')[['账单金额']].sum()
fn1.columns = ['60天账期的账单金额']

# 60天账期的入催金额
df4 = df3[(df3['账期'] == 60) & (df3['是否到期'] == 1) & (df3['历史逾期天数'] > 0)]
fn2 = df4.groupby('账单季度')[['未收金额2']].sum()
fn2.columns = ['60天账期的入催金额']

#90天账期的账单金额
df4 = df3[(df3['账期'] == 90) & (df3['是否到期'] == 1)]
fn3 = df4.groupby('账单季度')[['账单金额']].sum()
fn3.columns = ['90天账期的账单金额']

#90天账期的入催金额
df4 = df3[(df3['账期'] == 90) & (df3['是否到期'] == 1) & (df3['历史逾期天数'] > 0)]
fn4 = df4.groupby('账单季度')[['未收金额2']].sum()
fn4.columns = ['90天账期的入催金额']
```

```python
df_arr = [fn1, fn2, fn3, fn4]
final2 = pd.concat(df_arr, axis = 1)
final2['60天账期入催率'] = round(final2['60天账期的入催金额'] / final2['60天账期的账单金额'], 3)
final2['90天账期入催率'] = round(final2['90天账期的入催金额'] / final2['90天账期的账单金额'], 3)
final2
```

```python
line = (
    Line()
    .add_xaxis(list(final1.index.values.astype(str)))
    .add_yaxis(
        "60天账期入催率",
        list(final2['60天账期入催率']),
        yaxis_index = 0,
        color = "#675bba",
        label_opts = opts.LabelOpts(is_show=False),
    )
    .set_global_opts(title_opts = opts.TitleOpts(title="不同账期入催率"),)
    .add_xaxis(list(final1.index.values.astype(str)))
    .add_yaxis(
        "90天账期入催率",
        list(final2['90天账期入催率']),
        yaxis_index = 0,
        color = "#d14a61",
        label_opts = opts.LabelOpts(is_show=False),
    )
)
# line.render_notebook()
line.render('./file/不同账期入催率.html')
```

```python
# 不同预期天数收回情况
df6 = df3[(df3['未收金额2'] == 0) & (df3['是否到期'] == 1)].copy()

bin_arr = [-999,0,5,10,15,20,30,60,90,999]
label_arr = ['0','1-5','6-10','11-15','16-20','21-30','31-60','61-90','91+']
df6['历史逾期天数'] = pd.cut(df6['历史逾期天数'], bins = bin_arr, right = True, labels = label_arr)
final3 = df6.groupby('历史逾期天数')[['账期']].count()
final3.columns = ['回收账单数']
final3
```

```python
ydata = final3['回收账单数'].values.tolist()

bar = (
    Bar()
    .add_xaxis(list(final3.index.values.tolist()))
    .add_yaxis("收回账单数",ydata,yaxis_index=0,color="#675bba")
    .set_global_opts(
        title_opts=opts.TitleOpts(title="不同逾期天数的已收回账单数"),
    )
)
# bar.render_notebook()
bar.render('./file/不同逾期天数的已收回账单数.html')
```
