# random()函数
> 创建随机矩阵

```python
import numpy as np

# 创建3行4列的矩阵，并且数据类型为浮点数
random = np.random.rand(3, 4)
print(f'随机矩阵:{random}')
print(f'随机矩阵类型:{type(random)}')

# 创建3行4列随机二维整形数组，指定数据为-1 ～5
randomInt = np.random.randint(-1, 5, size=(3, 4))
print(f'随机矩阵:{randomInt}')
print(f'随机矩阵类型:{type(randomInt)}')

# 创建3行4列随机二维整形数组，指定数据为-1 ～5均匀样本值
randomUniform = np.random.uniform(-1, 5, size=(3, 4))
print(f'随机矩阵:{randomUniform}')
print(f'随机矩阵类型:{type(randomUniform)}')
```

# ndarray的数据类型
> 1、dtype参数，指定数组的数据类型，类型名+位数，如float64，int32

> 2、astype方法，转换数组的数据类型

```python
import numpy as np

# 初始化一个一个三行四列的数组，类型为float64、int32
zeros_float_arr = np.zeros((3, 4), dtype=np.float64)
print(f'zeros_float_arr:{zeros_float_arr}')
print(f'zeros_float_arr类型:{zeros_float_arr.dtype}')

# 将已有的数组数据类型转换为int32
zeros_int_arr = zeros_float_arr.astype(np.int32)
print(f'zeros_int_arr:{zeros_int_arr}')
print(f'zeros_int_arr类型：{zeros_int_arr.dtype}')
```

# 等比/等差数列
1、np.logspace()等比数列 np.logspace('开始点', '结束点', '元素个数')

2、np.linspace()等差数列

```python
import numpy as np

# 创建一个开始点为0，结束点为0，长度为10
gp1 = np.logspace(0 ,0 ,10)
print(f'等比数列-1:{gp1}')

# 创建一个开始点为0，结束点为9，长度为10
gp2 = np.logspace(0, 9, 10)
print(f'等比数列-2:{gp2}')

# 创建一个开始点为0，结束点为9， 长度为10，底数为2
gp3 = np.logspace(0, 9, 10, base=2)
print(f'等比数列-3:{gp3}')

# 创建一个开始点为1，结束点9，长度为10
ap = np.linspace(1, 9, 10)
print(f'等差数列-1:{ap}')

# 创建一个开始点0， 结束点9，长度为10
ap2 = np.linspace(0, 9, 10)
print(f'等差数列-2：{ap2}')

# 创建一个开始点1，结束点10，长度为10， 包含终止值
ap3 = np.linspace(1, 10, 10, endpoint=False)
print(f'等差数列-3：{ap3}')
```

# NnmPy内置函数

```python
import numpy as np

# 创建一个具有标准正态分布的序列
arr = np.random.randn(2, 3)
print(f'随机序列:{arr}')

print(f'随机序列向上取整:{np.ceil(arr)}')
print(f'随机序列向下取整:{np.floor(arr)}')
print(f'随机序列四舍五入:{np.rint(arr)}')
print(f'随机序列元素是否为空:{np.isnan(arr)}')
print(f'随机序列元素相乘:{np.multiply(arr, arr)}')
print(f'随机序列元素相除:{np.divide(arr, arr)}')
print(f'随机序列元素绝对值:{np.abs(arr)}')
print(f'随机序列元素判断:{np.where(arr > 1, 1, -1)}')
```

# 统计函数

```python
import numpy as np

# 生成一个元素不大于12的3x4的等差数列
arr = np.arange(16).reshape(4, 4)

print(f'等差数列:{arr}')
print(f'等差数列元素平均值:{np.mean(arr)}')
print(f'等差数列元素和:{np.sum(arr, axis = 0)}')
print(f'等差数列元素和:{np.sum(arr, axis = 1)}')
print(f'等差数列元素最大值:{np.max(arr)}')
print(f'等差数列元素最小值:{np.min(arr)}')
print(f'等差数列元素标准差:{np.std(arr)}')
print(f'等差数列元素方差:{np.var(arr)}')
print(f'等差数列最大值下标索引:{np.argmax(arr)}')
print(f'等差数列最小值下标索引:{np.argmin(arr)}')
print(f'等差数列元素累加和:{np.cumsum(arr)}')
print(f'等差数列元素累计乘积:{np.cumprod(arr)}')
```
