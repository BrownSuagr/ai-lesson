```python
# 从最大的N个值选取一个最小值，找到最小成本的高口碑电影
movies = pd.read_csv('../data/movie.csv')
```

```python
# 使用nlargest方法imdb分数最高100部
imdb_score_largest = movies.nlargest(100, 'imdb_score')

# 使用nsmallest获取预算最小的5部
budget_smallest = imdb_score_largest.nsmallest(5, 'budget')
budget_smallest
```

```python
# 通过排序获取每一组的最大值-找到每年IMDB评分最高的电影（sort_values()）
sorted_by_title_year = movies.sort_values(by = 'title_year', ascending = False)
sorted_by_title_year
```

```python
# 对两列数据进行排序
sorted_by_year_and_score = movies.sort_values(by = ['title_year','imdb_score'], ascending = False)
sorted_by_year_and_score
```

```python
# 使用drop_duplicates对数据进行去重
movies_duplicate_title = movies.drop_duplicates(subset = 'title_year')
movies_duplicate_title.head()
```

```python
# 多列排序
sorted_arr = ['movie_title', 'title_year', 'content_rating', 'budget']
sorted_movies = movies.sort_values(sorted_arr, ascending = [True, True, True, True])
sorted_movies.drop_duplicates(subset = ['title_year', 'content_rating'])
```

```python
# 加载LJ数据
house_data = pd.read_csv('../data/LJdata.csv')
house_data.head()
```

```python
# 替换原始列明
house_data.columns
```

```python
en_columns_arr = ['district', 'address', 'title', 'house_type', 'area', 'price', 'floor', 'build_time', 'direction', 'update_time', 'view_num', 'extra_info', 'link']
house_data.columns = en_columns_arr
```

```python
house_data.info()
```

```python
# 获取数据行列
print(house_data.shape)

# 获取数据描述
house_data.describe()
```

```python
# 找出数据中租金最高的和最低的房子
house_data.loc[house_data['price'] == house_data['price'].max()]
```

```python
house_data.sort_values(by = 'price', ascending = False).head(1)
```

```python
house_data.sort_values(by = 'price', ascending = True).head(1)
```

```python
# 找到最近新上的10套房子
house_data.sort_values(by = 'update_time', ascending = False).head(10)
```

```python
# 获取单列数据并去重
print(house_data['update_time'].unique())

# 获取带看人数的并均值
print(house_data['view_num'].mean())

# 带看人数中位数
print(house_data['view_num'].median())
```

```python
# 获取不同看房人数房源数量
group_by_view_num = house_data.groupby('view_num', as_index = False)['district'].count()
group_by_view_num
```

```python
group_by_view_num['district'].plot(kind = 'bar', figsize = (20, 10))
```

```python
print(house_data['price'].mean())   #平均值
print(house_data['price'].std())    #标准差
print(house_data['price'].median())  #中位数
```

```python
popular_direction = house_data.groupby('direction',as_index=False)[['view_num']].sum()
popular_direction[popular_direction['view_num']==popular_direction['view_num'].max()]
```

```python
# 设置正常显示汉字和负号
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 正常显示汉字
plt.rcParams['axes.unicode_minus'] = False # 正常显示负号


house_type_dis = house_data.groupby(['house_type']).count()
%matplotlib inline
house_type_dis['district'].plot(kind='bar') #柱状图
```

```python
# 获取最受欢迎的房型
group_by_view_num = house_data.groupby('house_type', as_index = False).agg({'view_num' : 'sum'})
group_by_view_num[group_by_view_num['view_num'] == group_by_view_num['view_num'].max()]
```

```python
# 获取房子的平均租金价格（元/平方米）
house_data['unit_price'] = house_data['price']/house_data['area']
mean_price = house_data['unit_price'].mean()
mean_price
```

```python
sum_view_num = house_data.groupby('address', as_index = False).sum()
sum_view_num[sum_view_num['view_num'] == sum_view_num['view_num'].max()]
```

```python
# 出租房源最多的小区
count_view_num = house_data.groupby('address', as_index = False).count()
count_view_num[count_view_num['view_num'] == count_view_num['view_num'].max()]
```
