```python
# 导入Pandas包
import pandas as pd

# 导入numpy包
import numpy as np
```

```python
# 创建数据
data = {
    'a': [10, 20, 30],
    'b': [20, 30, 40],
}
df = pd.DataFrame(data)
df
```

```python
# 创建一个自定义函数（单个参数）
def my_sq(x):
    """
    求平方
    :param x: 
    :return: 
    """
    return x ** 2
```

```python
df['square'] = df['b'].apply(my_sq)
df
```

```python
# 创建一个自定义函数（单个参数）
def my_power(base, index):
    """
    求平方
    :param x: 
    :return: 
    """
    return base ** index
```

```python
df['power_two'] = df['a'].apply(my_power, index=2)
df
```

```python
df['power_three'] = df['a'].apply(my_power, index=3)
df
```

```python
# DataFrame的apply方法
df.apply(my_sq)
```

```python
def my_avg(col):
    x = col[0]
    y = col[1]
    z = col[2]
    """
    编写函数计算列的平均值
    :param x: 
    :param y: 
    :param z: 
    :return: 平均值
    """
    return (x + y + z) / 3
```

```python
df.apply(my_avg)
```

```python
def my_avg_row(row):
    """
    按行处理求平均值
    :param row: 
    :return: 行平均值
    """
    x = row[0]
    y = row[1]
    return (x + y) / 2
```

```python
df.apply(my_avg_row, axis=1)
```

```python
# apply 使用案例

# 1、加载泰坦尼克号数据
titanic = pd.read_csv('../data/titanic.csv')
titanic.info()
# 数据891行、15列、age、embarked、deck、embark_town包含缺失值
```

```python
def count_missing(vector):
    """
    计算向量中缺省值的个数
    :param vector: 
    :return: 
    """
    # 获取向量中的为null值
    null_vector = pd.isnull(vector)
    # np函数求向量true值数量
    null_count = np.sum(null_vector)
    return null_count
```

```python
def prop_missing(vector):
    """
    计算向量中缺省值的占比
    :param vector: 
    :return: 
    """
    # 计算向量中null值个数
    num = count_missing(vector)
    # 获取向量个数
    size = vector.size
    return num / size
```

```python
def prop_complete(vector):
    """
    计算向量中非缺省值的占比
    :param vector: 
    :return: 
    """
    # 获取缺省值的占比
    pm = prop_missing(vector)
    return 1 - pm
```

```python
titanic.apply(count_missing)
```

```python
titanic.apply(prop_missing)
```

```python
titanic.apply(prop_complete)
```

```python
titanic.apply(count_missing, axis=1)
```

```python
titanic.apply(prop_missing, axis=1)
```

```python
titanic.apply(prop_complete, axis=1)
```

```python
titanic.apply(count_missing, axis=1).value_counts()
```

```python
# 向量化函数
data = {
    'a': [10, 20, 30],
    'b': [20, 30, 40]
}

df2 = pd.DataFrame(data)
df2
```

```python
def my_avg2(x, y):
    """
    计算平均值
    :param x: 
    :param y: 
    :return: 
    """
    if 20 == x:
        return np.NaN
    else:
        return (x + y) / 2
```

```python
# 1、将函数向量化
avg_mod_vector = np.vectorize(my_avg2)

# 2、对向量使用平均值函数
avg_mod_vector(df2['a'], df2['b'])
```

```python
@np.vectorize
def my_avg_decorator(x, y):
    """
    计算平均值
    :param x: 
    :param y: 
    :return: 
    """
    if 20 == x:
        return np.NaN
    else:
        return (x + y) / 2
```

```python
# 使用np.vectorize装饰器将函数向量化
my_avg_decorator(df2['a'], df2['b'])
```
