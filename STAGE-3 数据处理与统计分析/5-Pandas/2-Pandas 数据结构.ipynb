```python
series1 = pd.Series([1.1, 2.33, 3.999, 1.1])
series2 = pd.Series([4.77, 51.1, 699.01])

# 计算两个Series相关系数
series1.corr(series2)

# 计算两个Series协方差
series1.cov(series2)

# Series去重
series1.drop_duplicates()

# 判断两个Series是否相同
series1.equals(series2)

# 绘制Series直方图
series1.hist()

# 判断Series中是否包含某些值
series1.isin([1.1])

# 返回最大值
series1.max()

# Series最小值
series1.min()

# Series算数平均值
series1.mean()

# Series中位数
series1.median()

# Series众数
series1.mode()

# Series指定位置的分位数
series1.quantile()

# Series取代某些值
series1.replace(1.1, 1000.99)

# Series随机采样值
series1.sample()

# Series值排序
series1.sort_values()

# Series转为DataFrame
series1.to_frame()

# 去重返回数组
series1.unique()
```

```python
scientists = pd.read_csv('../data/scientists.csv')
scientists
```
