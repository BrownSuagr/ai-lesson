# 1、什么是条件概率、联合概率？
- 条件概率：表示事件A在另一个事件B发生的条件下的概率 P(A|B) = (P(B|A)P(A)/P(B))
- 联合概率：表示两个或者多个随机变量同时发生的概率 P(AB) = P(A)P(B|A), 特征条件独立性假设：P(AB) = P(A) P(B)

# 3、什么是朴素贝叶斯，其中朴素的含义是？
朴素贝叶斯：在特征条件独立假设加持下，可以将条件概率贝叶斯公式转化为：P(A|B) = (P(B|A)P(A)/P(B))= P(A)P(B)，其中这个特征条件独立假设被称为朴素

# 小结：
- 概率、联合概率、条件概率是概率模型计算中常用的计算公式
- 朴素贝叶斯中的朴素是指特征条件独立性假设
- 拉普拉斯平滑系数主要为了解决概率为0的问题
