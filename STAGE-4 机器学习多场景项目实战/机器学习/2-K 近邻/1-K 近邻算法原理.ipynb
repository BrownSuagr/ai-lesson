# 1、什么是KNN？

> KNN是一种非参数、有监督的学习分类器

# 2、KNN基本原理是什么？

> KNN 使用邻近度对单个数据点的分组进行分类或预测

# 3、KNN算法三要素是什么？
- 分类决策准测
- 距离度量
- K值选择

# 4、KNN算法步骤？
> 先确定K值，再计算距离，最后挑选K个最近的邻居进行投票
