```python
# 导入sklearn中鸢尾花的数据集
from sklearn.datasets import load_iris

# 导入sklearn中标准化包
from sklearn.preprocessing import StandardScaler

# 导入sklean中K临近算法
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 1、加载数鸢尾花数据集
iris_data  = load_iris()

# 2、对数据进行标准化
standard_scaler = StandardScaler()
standard_data = standard_scaler.fit_transform(iris_data.data)

# 3、对数据进行训练
knn_estimator = KNeighborsClassifier(n_neighbors=3)
knn_estimator.fit(standard_data, iris_data.target)

# 4、对数据进行预测
predict_result = knn_estimator.predict(standard_data)
print(predict_result)
```
