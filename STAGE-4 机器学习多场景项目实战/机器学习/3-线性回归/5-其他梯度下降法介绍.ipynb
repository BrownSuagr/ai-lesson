# 其他梯度下降法介绍
了解全梯度下降、随机梯度下降、小批量梯度下降、随机平均梯度下降的原理

---

# 1、全梯度下降算法（FGD）
- 概念：全梯度下降算法是每一次迭代使用全部的样本梯度值
- 优点：全梯度下降算法通常在数据集较小或计算资源充足时使用，因为它可以提供更稳定和准确的参数更新
- 缺点：FG每迭代一次权重都要计算全部样本误差，在样本数量较大的情况下效率偏低，且容易陷入局部最优解

# 2、批量梯度下降法
- 概念：是梯度下降最常用的形式，具体形式就是在更新参数时使用所有的样本来进行更新
- 优点：批量梯度下降每次迭代都使用整个数据集来计算梯度，
- 缺点：这可能导致计算量大，但可以确保每次更新都是基于全局信息进行的

# 3、随机梯度下降算法（SGD）
- 概念：随机梯度下降算法（SGD）每次迭代时, 随机选择并使用一个样本梯度值
- 优点：每轮的计算目标函数不再是全体样本误差，仅仅是单个样本误差，即每次带入计算的一个样本目标函数的梯度来更新权重，再取下个样本重复此过程，直到损失函数值停止或损失函数值小于某个可以容忍的阈值。

![SGD迭代形式](../img/SGD迭代形式.png)

- 缺点：在SG方法中，虽然避开了运算成本大的问题，但对于大数据训练而言，SG效果常不尽如人意，因为每一轮梯度更新都完全与上一轮的数据和梯度无关。

# 4、小批量梯度下降算法（mini-batch）
- 概念：每次迭代时, 随机选择并使用小批量的样本梯度值
- 优点：小批量梯度下降算法是FG和SG的折中方案,在一定程度上兼顾了以上两种方法的优点。

# 5、随机平均梯度下降算法（SAG）
- 概念：每次迭代时，随机选择一个样本的梯度值和以往样本的梯度值的均值
- 优点：随机平均梯度算法克服了这个问题，在内存中为每一个样本都维护一个旧的梯度，随机选择第i个样本来更新此样本的梯度，其他样本的梯度保持不变，然后求得所有梯度的平均值，进而更新了参数。

![SAG迭代形式](../img/SAG迭代形式.png)

# 小结
- 全梯度下降算法（FG）：在进行梯度下降迭代时，所有样本参与计算
- 随机梯度下降算法（SG）：在进行梯度下降迭代时，每次迭代只选取一个样本进行计算
- 小批量梯度下降算法（mini-batch）：在进行梯度下降迭代时，每次迭代只选取一部分样本进行计算
- 随机梯度下降算法（SAG）：在进行梯度下降迭代时，随机选择一个样本的梯度值和以往的梯度值的均值
