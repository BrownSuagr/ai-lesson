# 2、MSE（Mean Squared Error）

![均方误差](../img/均方误差.png)

# 3、RMSE（Root Mean Squared Error）
- 概念：均方根误差

![均方根误差](../img/均方根误差.png)

# 4、MAE、MSE、RMSE比较
- MAE、RMSE的优点：都能反映预测值和真实值之间的误差；MAE是真实值的平均误差，RMSE会将误差的数据点放大
- MAE、RMSE的缺点：MAE不能体现出误差大的数据点；RMSE虽然会放大误差大的数据点对指标的影响，但是对异常数据比较叫敏感

# 5、R-Squared 和 Adj.R²
![r-squared](../img/r-squared.png)

R²代表了预测值和真实值拟合的拟合程度，既考虑了预测值与真实值的差异，同时也兼顾了真实值的离散程度

- R²<0.5 → 弱拟合
- 0.5 ≤ R² ≤ 0.8 → 中度拟合
- R² > 0.8 强拟合

# 6、Adjusted R² (Adj.R²)
![adjusted-r](../img/adjusted-r.png)

作用：Adjusted R² 可以在R² 的基础上对模型是否过拟合提供更多的信息

# 小结：
- 常用的回归问题的评估方法：MAE、MSE、RMSE
- MAE、MSE、RMSE作用：RMSE会放大预测勿误差较大的样本影响；MAE对误差不敏感；RMSE对异常数据敏感
- R²和Adjusted R²：R²<0.5 → 弱拟合；0.5 ≤ R² ≤ 0.8 → 中度拟合； R² > 0.8 强拟合
- 不能单独通过R²的大小来判断模型好坏，可以通过配合Adjusted R² 和 多用不同的测试数据来修正R²的不足
