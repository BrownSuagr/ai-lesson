# 1、什么是信息增益率？
- 概念：信息增益率是指在选择某个属性进行数据集分裂后，数据集的不确定性（熵）的减少量
- 本质：信息增益率的本质是在信息增益的基础上乘以一个惩罚参数，特征个数较多时，惩罚参数较小，特征个数较少时，惩罚参数较大，惩罚参数：数据集D以特征A作为随机变量的熵的倒数

# 2、ID3和C4.5决策树优缺点对比


| 类型 | ID3                                                                                                                                | C4.5                                                                                                                                                      |
|:--:|:-----------------------------------------------------------------------------------------------------------------------------------|:----------------------------------------------------------------------------------------------------------------------------------------------------------|
| `优点` | 1、使用信息增益作为划分标准，‌计算简单，‌易于理解。‌<br/>2、适用于处理离散数据，‌对于分类问题有较好的表现。‌                                                                       | 1、分类规则利于理解，准确率比较高 <br/> 2、 C4.5算法核心思想史ID3算法，对ID3算法进行了相应的改进 <br/> 3、 C4.5使用的是信息增益比来选择特征，克服了ID3的不足 <br/>4、 可以处理离散性描述属性，也可以处理连续数值型属性 <br/>5、 能处理不完整的数据 <br/> |
| `缺点` | 1、不能处理具有连续值的属性 <br/>2、不能处理属性具有缺失值的样本 <br/>3、算法会生成很深的树，容易产生过拟合的现象 <br/>4、算法一般会优先选择较多属性的特征，因为属性较多的特征会有相对较大的信息增益，但这里的属性并不一定是最优的 <br/> | 1、在构造过程中，需要对数据集进行多次的顺序扫描和排序，导致算法低效<br/>2、C4.5只适应于能够驻留内存的数据集，当数据集非常大时，程序无法运行                                                                               |

`备注`：无论是ID3还是C4.5最好在小数据集上使用，当特征值很多是最好还是使用C4.5算法
