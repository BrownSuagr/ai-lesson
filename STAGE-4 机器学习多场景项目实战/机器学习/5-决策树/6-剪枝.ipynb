
# 剪枝学习目标：
- 知道什么是剪枝，为什么要剪枝？
- 知道常用的剪枝方法

---


# 1、什么是剪枝？
剪枝（pruning）是对决策树算法过拟合的主要手段

# 2、为什么要进行剪枝？
决策树是充分考虑所有的数据点形成的复杂树，有可能出现过拟合的情况，决策树越是复杂，过拟合的程度会越高，而剪枝是对分裂前后分类误差相差不大的子树进行剪枝，能够降低决策树的复杂程度，降低拟合的过程。

# 3、如何进行剪枝？
剪枝有两种方案：预剪枝（pre-pruning）和后剪枝（post-pruning）
- 先剪枝：先剪枝是提前结束决策树的生长，跟决策树停止生长的方法是一致的
- 后剪枝：是在决策树生长完成后在进行剪枝的过程

# 4、常见的剪枝方法有哪些，有什么特点？
决策树的剪枝策略有两种：预剪枝（pre-pruning）和后剪枝（post-pruning）
- 预剪枝：是在决策树生长过程中，对每个节点在划分前先进行估计，若当前的节点划分不能带来决策树泛化性能的提升，则停止划分并将当前节点标记为叶节点
- 后剪枝：后剪枝是先从训练集中生成一棵完整的决策树，然后自底向上对非叶子节点进行考察，拖将该节点子树替换为叶节点能带来决策树泛化能力的提升，则将该子树替换为叶节点




# 5、剪枝技术优缺点比对
预剪枝
- 优点：预剪枝使决策树喝多分支没有展开，不单降低了过拟合的风险，还显著减少了决策树训练和测试时间的开销
- 缺点：有些分支当前的划分能力虽然不能提升泛化能力，甚至会导致泛化能力降低，但是在其基础上进行后续的划分有可能会导致性能的明显提高；预剪枝决策树可能带来欠拟合的风险
后剪枝
- 优点：比预剪枝保留了更多的分支，一般情况下，后剪枝决策树欠拟合的风险更小，泛化性能往往优于预剪枝
- 缺点：后剪枝过程是在完全生成决策树后进行，并要自底向上的对树中的非叶子节点进行逐一的考察，因此在训练时间开销和比未剪枝的决策树和预剪枝的决策树都要大的多




# 小结：
- 剪枝是决策树对付过拟合的手段
- 剪枝主要有预剪枝和后剪枝两种手段
