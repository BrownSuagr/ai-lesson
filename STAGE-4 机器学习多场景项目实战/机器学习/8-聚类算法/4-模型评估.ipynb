# 1、什么是SSE（误差平方和）？
定义：通常是评估一个统计模型的拟合程度。它指的是模型预测值与实际观测值之间差异的平方和，
特性：在回归分析中SSE是衡量模型预测准确度的一个总要指标，SSE越小表示模型对数据拟合越好。

# 2、什么是“肘方法”（Elbow method），怎么确定K值？

![肘方法](../img/肘方法.png)

定义：是一种用于确定聚类分析中最佳聚类数量的方法，
确定K值的方法：
- 对于N个点的数据集，迭代计算K从1～n，每次聚类完成后计算每个点到其所属的簇中心的距离的平方和
- 平方和是会逐渐变小的，直到k等于n时平方和为0，因为每个点都是它所在的簇中心本身
- 在这个平方变化过程中，会出现一个拐点即“肘”点，`下降率突然变平缓时即认为是最佳的K值`
`备注：`“肘形”判据对于训练停止也有效，在某个类别增加不能带来更多回报时，即我们认为可以停止某个类别的增加。


# 3、什么是轮廓系数法（Silhouette Coefficient）？
说明：结合了聚类的凝聚度（Cohesion）和分类度（Separation）
定义：是一种用于评估聚类好坏的统计量，通过测量每个数据点与其所形成的聚类内部成员的紧密度以及与其最近的其他聚类的距离来评估聚类的有效性
- 轮廓系数（SC系数）的范围是：[-1, 1]
- 接近-1表示聚类效果差
- 接近1表示聚类效果好
- 在0附近表示数据位于两个聚类的边界上

# 小结
- SSE：误差平方和的值越小越好
- 肘方法：下降率突然变缓即认为是最佳的K值
- SC系数：取值为[-1, 1],其值越大越
