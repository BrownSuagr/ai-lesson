# K-Means API学习目标：
- 掌握聚类算法API的使用方法
```python
estimator.fit(x)
estimator.predict(x)
# 相当于先调用了fit 后在调用predict函数
estimator.fit_predict(x)
```

--- 

# 小结
- API使用
```python
sklearn.cluster.KMeans(n_clusters=8)
# 参数: n_clusters:开始的聚类中心数量
# 方法: estimator.fit_predict(x)
```
- 计算聚类中心并预测每个样本属于那个类别，相当于先调用fit(x),然后在调用predict(x)

````python
# 导入matplotlib绘图包
import matplotlib.pyplot as plt
# 导入数据集产生包
# from sklearn.datasets.samples_generator import make_blobs
from sklearn.datasets import make_blobs
# 导入K-Means包
from sklearn.cluster import KMeans
# 导入轮廓系数包
from sklearn.metrics import silhouette_score
````

````python
# 1、创建样本数据集
center_arr = [[-1, -1], [0, 0], [1, 1], [2, 2]]
cluster_std_arr = [0.4, 0.2, 0.2, 0.2]
X, y = make_blobs(n_samples = 10000, n_features = 2, centers = center_arr, cluster_std = cluster_std_arr, random_state = 0)

print(X)
print(y)
# 2、数据集可视化
plt.scatter(X[:, 0], X[:, 1], marker = 'o')
plt.show()
  
# 3、使用silhouette_score轮廓系数进行评估
# 参数说明：
# n_clusters：要分成的簇的数量（K值）
# init：初始化中心点的方法，可以是k--、means++、random、数组
# n_init:使用不同的初始化种子运行K-Means的次数，以提高结果稳定性
# max_iter:算法的最大迭代次数
# tol：收敛容忍度，即中心点小于某个值就停止迭代

y_pred = KMeans(n_clusters = 2, random_state = 9).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.show()

# 4、打印轮廓系数评分
silhouette_avg = silhouette_score(X, y_pred)
print('轮廓系数评分：', silhouette_avg)
````

````python
# 1、创建样本数据集
center_arr = [[-1.5, -2], [0.5, 0.8], [1.2, 0.8], [2.1, 1.8]]
cluster_std_arr = [0.4, 0.2, 0.2, 0.2]
X, y = make_blobs(n_samples = 10000, n_features = 2, centers = center_arr, cluster_std = cluster_std_arr, random_state = 0)

print(X)
print(y)
# 2、数据集可视化
plt.scatter(X[:, 0], X[:, 1], marker = 'o')
plt.show()

# 3、使用silhouette_score轮廓系数进行评估
# 参数说明：
# n_clusters：要分成的簇的数量（K值）
# init：初始化中心点的方法，可以是k--、means++、random、数组
# n_init:使用不同的初始化种子运行K-Means的次数，以提高结果稳定性
# max_iter:算法的最大迭代次数
# tol：收敛容忍度，即中心点小于某个值就停止迭代

y_pred = KMeans(n_clusters = 4, random_state = 20).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.show()

# 4、打印轮廓系数评分
silhouette_avg = silhouette_score(X, y_pred)
print('轮廓系数评分：', silhouette_avg)
````
