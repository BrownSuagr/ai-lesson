```python
# 导入pandas包
import pandas as pd

data_set = pd.read_csv('../file/data/customers.csv')
data_set.head()
```

```python
# 导入无监督学习聚类算法KMeans
from sklearn.cluster import KMeans

# 1、截取第四五列
X = data_set.iloc[:, [3, 4]].values

# 2、使用聚类算法KMeans(5个聚类算法中心，算法使用k-means++ 随机生成数初始状态42)
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
'''
    K-means和K-means++区别（选取聚类中心思想不一致）
    1、K-means：随机选取数据集中k个点作为聚类中心
    2、K-means++：假设已经选取了n个初始聚类中心（0<n<k）,在选取n+1个聚类中心时，距离当前n个聚类中心越远的点会有更高的概率被选取为n+1个聚类中心，在选取第一个聚类中心（n=1）时同样随机的方法
'''

# 3、训练并产生预测结果
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
# 导入绘图包
import matplotlib.pyplot as plt

color_arr = ['red', 'blue', 'green', 'cyan', 'magenta', 'black']
label_arr = ['Standard', 'Traditional', 'Normal', 'Youth', 'TA', 'Centroids']
center_arr = kmeans.cluster_centers_

# 4、绘图展示
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = color_arr[0], label = label_arr[0])
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = color_arr[1], label = label_arr[1])
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = color_arr[2], label = label_arr[2])
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = color_arr[3], label = label_arr[3])
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = color_arr[4], label = label_arr[4])
plt.scatter(center_arr[:, 0], center_arr[:, 1],       s = 300, c = color_arr[5], label = label_arr[5])
# 标题
plt.title('Clusters of customers')
# x轴标签
plt.xlabel('Annual Income(K$)')
# y轴标签
plt.ylabel('Spending Score(1~100)')
plt.legend()
plt.show()
```

```python
# 导入绘图包
import matplotlib.pyplot as plt

# 5、对聚类进行评估（肘方法）
wcss_arr = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    # inertia簇内误差平方和
    wcss_arr.append(kmeans.inertia_)

print('肘方法数据：', wcss_arr)
plt.plot(range(1, 11), wcss_arr)
# 标题
plt.title('The Elbow Method')
# x轴标签
plt.xlabel('Number of clusters')
# y轴标签
plt.ylabel('WCSS')
plt.show()
```

# 总结
- K-Means方法特点：K均值是聚类中最常见的方法之一，给予点与点之间的距离的相似度来计算最佳的类别归属
- K均值需要注意点：
    - 数据异常值：数据中的异常值能明显改变不同点之间的距离的相似度，并且这种影响是非常显著的，因此基于距离相似度的判别模式下，异常值的处理必不可少
    - 数据异常量纲：不同的纬度和变量之间，如果存在数值或量纲差异，那么在做距离之前先要将变量进行归一化或标准化。
- 数据量过大不适合使用K均值原因：
    - K-Means在算法稳定性、效率、准确度（相对于真实标签）表现非常好，并且在应对大数据量依然如此，它的算法时间复杂度上界O(n*k*t)，其中n是样本数量、k是划分聚类数、t迭代次数
    - 当聚类数和迭代次数不变时，K均值的算法消耗时间只跟样本量有关，因此会呈现线性增长趋势，当真正面对海量数据时，使用K均值算法将面临严重的记过延迟，尤其是当K均值被用作实时性或准实时性的数据预测处理、分析和建模时、这种瓶颈效应尤其明显。
