# **学习目标**
- 理解什么是欠拟合
- 理解什么是过拟合
- 理解什么是泛化能力


# 1、什么是欠拟合，有哪些解决办法？

概念：模型在训练集表现不佳，在测试集也表现不佳

解决办法：
- 添加其他特征项
- 添加多项式特征
- 减少正则化参数

# 2、什么是过拟合，产生原因有哪些？有哪些解决办法？

概念：在训练集表现不错，在测试集表现不佳

产生原因：模型太复杂、数据不纯、测试集数据量太少

 
解决办法：

 - 重新对数据清洗
 - 增加测试集训练数据量
 - 采用正则化方法对模型参数施加惩罚
 - 使用Dropout方法

# 4、什么是泛化能力？
模型对未知数据的预测能力
