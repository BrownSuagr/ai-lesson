# 什么是监督学习，及常见模型有哪些？
> 训练样本的期望输出已知，学习目的是对新样本的输出进行预测，典型的任务包括分类和回归，常见模型有Linear Regression/Logistic Regression/ SVM/ Neural Network

```python
# 什么是无监督学习，常见的模型有哪些？
# 对没有类别标记的样本进行学习，学习目的通常是发现数据内在结构，典型任务是聚类和降维。
```

```python
# 什么是半监督学习（Semi-supervised learning）？
# 学习器自行利用少量的具有标记信息的样本和大量没有标记的样本进行学习的框架。
```

```python
# 什么是强化学习（reinforcement learning），常见模型有哪些？
# 一种从环境状态到动作映射的学习，目标是使动作从环境中获得的累积奖赏值最大。
```
