```python
# 人工智能的应用方向有哪些？
# 无人驾驶、智能翻译、图像识别、数据挖掘、医疗预测
```

```python
# 人工智能重要发展节点？
# 有三个重要节点：
# 第一次浪潮：1956年是人工智能发展元年以符号主义流派为主的专家系统为主导，主要事件有：图灵设计国际象棋程序、IBM 亚瑟·塞米尔战胜人类跳棋高手
# 第二次浪潮：80年代初以系统主义流派为主使用统计模型解决问题，主要事件有：Vapnik提出SVM IBM深蓝战胜卡斯帕罗夫
# 第三次浪潮：10年以后以神经网络和深度学习为主，主要事件有：Hinton，AlphaGO战胜围棋高手李世石
```

```python
# 人工智能、机器学习、深度学习关系是什么？
# 机器学习是人工智能的一个方向，深度学习是机器学习的一个技术路线
```

```python
# 为什么要学习机器学习？
# 1、可以提取数据中有价值的部分
# 2、从数据分析的4层角度理解，机器学习帮助我们实现预测和规范性分析
```

```python
# 小节
# 1、人工智能在无人驾驶、图像识别、医疗智能诊断、数据挖掘方面广泛应用
# 2、1956 年被称为人工智能元年，发展至今经历了三次浪潮
# 3、机器学习是人工智能的一个实现途径,深度学习是机器学习的一个方法发展而来
# 4、学习机器学习可以对数据进行价值化提取，帮助我们实现预测和规范性
```
