# 1、什么是聚类算法？
聚类算法是一种典型的无监督学习算法，主要是将相似的样本自动归到一个类别中，而且样本之间的相似性使用欧氏距离测算



# 2、聚类算法有哪些应用场景？
- 用户画像、广告推荐、Data Segmentation、搜索引擎流量推荐、恶意流量识别、
- 基于位置信息的商业推送、新闻聚类、筛选排序
- 图像分割、降维、识别、离群点检测
- 信用卡异常消费
- 发掘相同功能的基因片段


# 3、聚类算法和分类算法有哪些区别？

# 小结：
- 聚类算法的定义：
    - 是一种典型的无监督学习算法
    - 主要用于将相似的样本自动归到一个类别
    - 计算样本和样本之间的相似性，一般使用欧氏距离
- 聚类和分类算法有哪些区别：聚类算法是无监督学习，分类算法属于监督学习算法
