```python
%%sql
-- 核心指标计算
select date                    as 日期,
       sum(liulan)             as 页面浏览次数,
       sum(buyer)              as 卖家访问次数,
       sum(on_sell)            as 平均在售商品数,
       sum(refund_ratio) / 100 as 退款率,
       sum(sales_total)        as 销量,
       sum(gmv_total)          as 销售额
from ds_sales_all
group by date
order by date desc
limit 7;
```
