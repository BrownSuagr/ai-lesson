# 1、什么是损失函数？
- 概念：用来衡量机器学习模型性能的函数,损失函数可以计算预测值与真实值之间的误差（用一个实数来表示），误差越小说明模型性能越好
- 作用：确定损失函数之后， 我们通过求解损失函数的极小值来确定机器学习模型中的参数



# 2、什么是平方损失？
> 平方损失函数就是对theta的极大似然估计

# 4、总结
- 损失函数在训练阶段能够指导模型的优化方向，在测试阶段能够用于评估模型的优劣。
- 线性回归使用平方损失
- 正规方程是线性回归的一种优化方法

```python
# 导入numpy包
import numpy as np
# 导入线性回归包
from sklearn.linear_model import LinearRegression

# 1、获取特征值和目标值
x = np.mat(
    [[80, 86],
     [82, 80],
     [85, 78],
     [90, 90],
     [86, 82],
     [82, 90],
     [78, 80],
     [92, 94]]
)

y = np.mat([84.2, 80.6, 80.1, 90, 83.2, 87.6, 79.4, 93.4]).transpose()

# 2、给特征列加一
ones_arr = np.ones([len(x), 1])
x = np.hstack([ones_arr, x])

# 3、使用正规方程计算w，b
w = ((x.transpose() * x)** -1) * x.transpose() * y
print(w)
print(w[0][0])
print(w[1])
print(w[1][0])
print(w[2])
print(w[2][0])
print('sss', w[0][0], w[1][0], w[2][0])
print('[%.1f %.1f %.1f]' % (w[0][0], w[1][0], w[2][0]))

# 4、使用线性回归求解
estimator = LinearRegression(fit_intercept=False)
estimator.fit(x, y)
print(estimator.coef_[0])
```
