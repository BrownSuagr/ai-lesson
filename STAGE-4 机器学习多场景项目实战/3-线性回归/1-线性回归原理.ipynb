# 2、什么是线性回归？
- 线性回归（Linear Regression）是利用回归方程（函数）对一个或者多个自变量（特征值）和因变量（目标值）之间进行建模的一个一种分析方式
- 特点：只有一个特征的情况称为单变量回归，多个特征值的属于多元回归

# 5、小结
- 线性回归公式
- 线性回归公式中，w、b 代表模型参数
- LinearRegression.fit 表示模型训练函数
- LinearRegression.predict 表示模型预测函数
- 线性回归模型的目标：通过学习得到线性方程的这两个权值，如在y=kx+b中，得到k和b两个权值，并用这个方程解释变量和目标变量之间的关系。




````python
# 导入线性回归包
from sklearn.linear_model import LinearRegression

# 1、加载数据
x = [[80, 86],
     [82, 80],
     [85, 78],
     [90, 90],
     [86, 82],
     [82, 90],
     [78, 80],
     [92, 94]]

y = [84.2, 80.6, 80.1, 90, 83.2, 87.6, 79.4, 93.4]

# 2、创建线性回归的对象
estimator = LinearRegression()

# 3、调用fit方法进行训练
estimator.fit(x, y)

# 4、打印线性回归回归系数
print(estimator.coef_)

# 5、对未知样本进行预测
simple = [[100, 80]]
estimator.predict(simple)
````
