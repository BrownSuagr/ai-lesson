# 1、什么是回归决策树？
概念："回归决策树"是一种机器学习算法，它用于解决回归问题，即预测一个连续的数值

# 2、CART回归树和CART分类树有什么不同？
- CART分类树预测输出是一个离散值，CART回归树预测输出是一个连续值
- CART分类树使用基尼指数作为划分、构建树依据，CART回归树使用平方损失
- 分类树使用叶子节点里出现更多次的类别作为预测类别，回归树则采用叶子结点里面均值作为预测输出

# 3、回归决策树的构建过程？
- 选择第一个特征，将该特征的值进行排序，取相邻点计算均值作为待划分点
- 根据所有的划分点，将数据集分成两部分：R1，R2
- R1和R2两部分的平方损失相加作为该切分点平方损失
- 取最小的平方损失的划分点，作为当前特征的划分点
- 一次计算其他特征的最有划分点，以及该划分点对应的损失值
- 在所有特征的划分点中，选择出最小平方损失的划分点，作为当前树的分裂点

# 小结：
- 回归决策树使用平方损失作为分裂增益计算指标
- 回归决策树是二叉树
