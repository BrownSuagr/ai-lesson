# 什么是高斯核支持向量机？
是一种机器学习算法，属于支持向量机（SVM）的一个变种。支持向量机是一种监督学习模型，用于分类和回归分析，其中“高斯核”是指：径向基函数RBF

# 1、什么是SVM？
概念：支持向量机是一种监督学习算法，主要用于分类和回归分析，是一种强大的机器学习模型，能够处理线性和非线性问题，并且对于高维的数据集特别有效

# 2、支持向量机有哪些分类？
![SVM分类](../img/SVM分类.png)

- 线性支持向量机（软间隔支持向量机）
- 线性可分支持向量机（硬间隔支持向量机）
- 非线性支持向量机
