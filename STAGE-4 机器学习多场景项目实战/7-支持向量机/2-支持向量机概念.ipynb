# 1、软间隔和硬间隔的区别？
- 硬间隔是指让所有样本都不在最大间隔之间，并为雨争取的一边，这就是硬间隔，如果出现异常值，或者样本不能线性可分，此时硬间隔无法实现
- 软间隔是指可以容忍一部分样本在最大间隔之内，甚至于在错误的一边，相对来说软间隔用在一些线性不可分的场景

# 小结：
- 支持向量机可以用于回归问题，也可以用户分类问题
- 硬间隔指：样本必须全部分类正确；软间隔值：允许部分样本间隔违例
- 惩罚参数C用于惩罚间隔违例样本带来的损失，C越大越不能容忍间隔违例，C越小越能容忍间隔违例
