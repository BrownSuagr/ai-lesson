# SVM的核方法和损失函数学习目标
- 知道常用的核函数
- 知道SVM的损失函数

---

# 1、什么是核函数，常见的核函数有哪些？
概念： 它允许算法在高维空间中进行有效的计算，而无需显式地映射数据到高维空间。核方法的核心思想是利用核函数来计算数据点在高维空间中的内积，从而实现数据的非线性转换。

![核函数](../img/核函数.png)

# 2、什么是损失函数？
概念： 它是一个衡量模型预测值与实际值之间差异的函数

# 3、常见损失函数包括哪些？

![常见损失函数](../img/常见损失函数.png)


# 小结：
- 常见的核方法： 线性核、多项式核、高斯核
- 损失函数包括：0-1损失、Hinge（合叶）损失、Logistic损失
